```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae,median_absolute_error as mee,classification_report as cr,accuracy_score as ac,r2_score as r2,mean_squared_error as mse
```

# Classification

```python
from sklearn import datasets 
wine = datasets.load_wine()
```

```python
x=wine.data
x
```

```python
y=wine.target
y
```

```python
print(x.shape)
print(y.shape)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
plt.plot(x,y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
print(st_x)
```

```python
x_train= st_x.fit_transform(x_train) 
print(x_train)
```

```python
x_test= st_x.transform(x_test)  
print(x_test)
```

```python
clf=svm.SVC(kernel='rbf',C=1,gamma=0.1)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pre=clf.predict(x_test)
y_pre
```

```python
ac(y_test,y_pre)*100
```

```python
from sklearn.metrics import confusion_matrix  
confusion_matrix(y_test, y_pre) 
```

```python
print(cr(y_test,y_pre))
```

```python
print(y_test)
print(y_pre)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(clf, wine.data, wine.target, cv=10)*100
```

```python
plt.plot(y_test)
```

```python
plt.plot(y_pre)
```

```python
plt.plot(y_test,y_pre)
```

# Regression

```python
from sklearn import datasets 
bos = datasets.load_boston()
```

```python
x=bos.data
x
```

```python
y=bos.target
y
```

```python
bos.feature_names
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
print(st_x)
```

```python
x_train= st_x.fit_transform(x_train) 
print(x_train)
```

```python
x_test= st_x.transform(x_test)  
print(x_test)
```

```python
clf=svm.SVR(kernel='linear',C=1,gamma=0.1)
```

```python
clf.fit(x_train,y_train)
```

```python
y_pre=clf.predict(x_test)
y_pre
```

```python
clf.score(x_train,y_train)*100
```

```python
clf.score(x_test,y_test)*100
```

```python
mae(y_test,y_pre)*100
```

```python
mee(y_test,y_pre)*100
```

```python
mse(y_test,y_pre)*100
```

```python
r2(y_test,y_pre)*100
```

```python
print("Coeff",clf.coef_.sum())
print("Int",clf.intercept_)
```

```python
print(y_test)
print(y_pre)
```

```python
plt.plot(y_test,y_pre)
```
