```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae,median_absolute_error as mee,classification_report as cr,accuracy_score as ac,r2_score as r2,mean_squared_error as mse
```

# Classification

```python
from sklearn import datasets 
br = datasets.load_breast_cancer()
```

```python
x=br.data
x
```

```python
y=br.target
y
```

```python
print(x.shape)
print(y.shape)
```

```python
br.feature_names
```

```python
br.target_names
```

```python
plt.plot(x,y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
print(st_x)
```

```python
x_train= st_x.fit_transform(x_train) 
print(x_train)
```

```python
x_test= st_x.transform(x_test)  
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
```

```python
clf=KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
                     metric_params=None, n_jobs=None, n_neighbors=5, p=2,
                     weights='uniform')
```

```python
clf.fit(x_train,y_train)
```

```python
y_pre=clf.predict(x_test)
y_pre
```

```python
ac(y_test,y_pre)*100
```

```python
from sklearn.metrics import confusion_matrix  
confusion_matrix(y_test, y_pre)  
```

```python
print(cr(y_test,y_pre))
```

```python
print(y_test)
print(y_pre)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(clf, br.data, br.target, cv=10)*100
```

```python
plt.plot(y_test)
```

```python
plt.plot(y_pre)
```

```python
plt.plot(y_test,y_pre)
```

# Regression

```python
from sklearn import datasets 
bos = datasets.load_boston()
```

```python
x=bos.data
x
```

```python
y=bos.target
y
```

```python
bos.feature_names
```

```python
print(x.shape)
print(y.shape)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=100)
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
print(st_x)
```

```python
x_train= st_x.fit_transform(x_train) 
print(x_train)
```

```python
x_test= st_x.transform(x_test)  
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
clf=KNeighborsRegressor()
```

```python
clf.fit(x_train,y_train)
```

```python
y_pre=clf.predict(x_test)
y_pre
```

```python
clf.score(x_train,y_train)*100
```

```python
clf.score(x_test,y_test)*100
```

```python
mae(y_test,y_pre)*100
```

```python
mee(y_test,y_pre)*100
```

```python
mse(y_test,y_pre)
```

```python
r2(y_test,y_pre)*100
```

```python
print(y_test)
print(y_pre)
```
